(a) What is the average annual log return over the data span?

To calculate the average annual log return, we need to first calculate the total log return over the data span and then divide it by the number of years in the data span.

The formula for log return is:

$log_{ret} = log(P_t/P_{t-1})$

where $P_t$ is the price of the stock at time $t$ and $P_{t-1}$ is the price of the stock at time $t-1$.

Using the data provided in Exercise 1.2, we can calculate the log return for each month using the formula above. Then we can add up all the log returns for each year to get the total log return for that year.

Next, we can divide the total log return by the number of years in the data span to get the average annual log return.

```python
import pandas as pd
import yfinance as yf
from pandas_datareader import data 
import numpy as np
```

```python
df = yf.download('^GSPC', start='1975-01-01', end='2003-01-01', interval='1mo')

df
```

```python
# Reset index to add Date to DF 

df1=df.reset_index()

df1
```

```python
# Calculate the monthly log returns
df1['Log Return'] = np.log(df1['Adj Close']/df1['Adj Close'].shift(1))

df1
```

```python
# Group the log returns by year and sum them up

yearly_log_ret = df1.groupby(pd.Grouper(key='Date', freq='Y'))['Log Return'].sum()

yearly_log_ret
```

```python
# Calculate the average annual log return 

average_annual_log_ret= yearly_log_ret.mean() 

average_annual_log_ret
```

```python
print("The average annual log return over the data span is {:.2%}".format(average_annual_log_ret))
```

(b) Assume that there were no transaction costs. If one invested $1.00 on the
S&P composite index at the beginning of 1985, what was the value of the
investment at the end of 2003?

```python
df1['cumulative_return'] = (1 + df1['Log Return']).cumprod()

df1
```

```python
# Find the cumulative return at the end of 2003

cum_return_2003 = df1.loc[df1['Date'] == '2002-12-01', 'cumulative_return'].values[0]

cum_return_2003
```

```python
# Calculate the value of the investment at the end of 2003
inv_value_2003 = cum_return_2003 * 1.00

inv_value_2003
```
